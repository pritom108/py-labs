**Matplotlib Line Styles**

Matplotlib provides several line styles to customize plots. Here is a complete list:

| Line Style | Description       |
|------------|-----------------|
| `'-'`      | Solid line       |
| `'--'`     | Dashed line      |
| `'-.'`     | Dash-dot line    |
| `':'`      | Dotted line      |
| `'None'`   | No line          |
| `' '`      | No line          |
| `''`       | No line          |

```python
x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x,y, marker="*",
              markersize=20,
              markerfacecolor="#1cd3fc",
              markeredgecolor="#1cd3fc",
              linestyle="dashed",
              linewidth=4,
              color="#1c5bfc",)

plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([19, 21, 38, 2])

plt.plot(x,y1, marker="*",
              markersize=20,
              markerfacecolor="#1cd3fc",
              markeredgecolor="#1cd3fc",
              linestyle="dashed",
              linewidth=4,
              color="#1c5bfc")

plt.plot(x,y2, marker=".",
              markersize=20,
              markerfacecolor="#1cd3fc",
              markeredgecolor="#1cd3fc",
              linestyle="solid",
              linewidth=4,
              color="#1c5bfc",)

plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y3 = np.array([15, 25, 30, 20])
y4 = np.array([19, 21, 38, 2])

line_style = dict(marker="*",
                markersize=20,
                markerfacecolor="#1cd3fc",
                markeredgecolor="#1cd3fc",
                linestyle="dashed",
                linewidth=4,
                color="#1c5bfc")

plt.plot(x,y3, **line_style)

plt.plot(x,y4,**line_style)

plt.show()
```

# Bar Charts

**Matplotlib `plt.bar()` Important Parameters**

`plt.bar()` is used to create vertical bar charts. Here are the key parameters:

- **x** : array-like  
  The x coordinates of the bars (categories).

- **height** : array-like  
  The height(s) of the bars (values).

- **width** : float, default 0.8  
  The width of the bars.

- **bottom** : float or array-like, default 0  
  The y-coordinate(s) of the bottom of the bars.

- **align** : {'center', 'edge'}, default 'center'  
  Alignment of bars relative to `x`.  
  `'center'`  bars are centered on x  
  `'edge'`  bars start at x

- **color** : color or array-like, optional  
  Bar fill color(s).

- **edgecolor** : color or array-like, optional  
  Color of bar edges.

- **linewidth** : float, optional  
  Width of the bar edges.

- **tick_label** : list of str, optional  
  Labels for the bars.

- **xerr**, **yerr** : scalar or array-like, optional  
  Error bars for x or y values.

- **ecolor** : color, optional  
  Color of the error bars.

- **log** : bool, default False  
  Set True to use log scaling on y-axis.

- **orientation** : {'vertical', 'horizontal'}, default 'vertical'  
  Orientation of bars. Use `'horizontal'` for horizontal bars.

- **hatch** : str, optional  
  Pattern to fill the bars (e.g., `'/'`, `'x'`, `'\\'`).

- **label** : str, optional  
  Label for the bars (used in legends).

**Example:**



```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(
    categories, values, 
    width=0.5, 
    color='skyblue', 
    edgecolor='black', 
    linewidth=1.5, 
    tick_label=['Cat A', 'Cat B', 'Cat C', 'Cat D'],
    label='Values'
)
plt.legend()
plt.show()
```

```python
categories = np.array(["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5 ,3,1])


plt.bar(categories,values)

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

# Pie Charts

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
colors= ["red","yellow","blue","green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",
                colors=colors,
                explode=[0,0,0,0.1],
                shadow=True,
                startangle=180)

plt.show()
```

# Scatter Graphs

```python
x= [0,1,1,2,3,4,5,6,7,7,8]
y= [55,60,65,62,68,70,75,78,82,85,87]

plt.scatter(x,y, color="skyblue",
                alpha=0.5,
                s=200)

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.show()
```

```python
x1= np.array([0,1,1,2,3,4,5,6,7,7,8])
y1= np.array([55,60,65,62,68,70,75,78,82,85,87])

x2= np.array([0,1,2,2,3,4,5,6,7,7,8])
y2= np.array([50,63,55,67,68,74,75,78,89,87,80])

plt.scatter(x1,y1, color="skyblue",
                    alpha=0.5,
                    s=200,
                    label="Class A")

plt.scatter(x2,y2, color="red",
                    alpha=0.5,
                    s=200,
                    label="Class B")

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.legend()

plt.show()
```

# Histograms

```python
scores = np.random.normal(loc=80, scale=10, size=100)

plt.hist(scores)

plt.show()
```

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores,0,100)

plt.hist(scores)

plt.show()
```

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores,0,100)

plt.hist(scores, bins=10,
                  color="lightgreen",
                  edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of Students")

plt.show()
```

```python
x = np.array([1,2,3,4,5])

figure, axes = plt.subplots(2,2)

axes[0, 0].plot(x,x*2, color="red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x,x**2, color="blue")
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x,x**3, color="green")
axes[1, 0].set_title("x**3")

axes[1, 1].plot(x,x**4, color="purple")
axes[1, 1].set_title("x**4")

plt.show()
```
